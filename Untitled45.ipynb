```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

#  
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
diabetes_df['target'] = diabetes.target

# 
y = diabetes_df['target']
X = diabetes_df.drop(['target'], axis=1)

#     
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)

#  
model = Sequential()

#    
model.add(Dense(150, input_dim=10, activation='relu'))

#  
model.add(Dense(1, activation='relu'))

#  
model.compile(loss='mse', optimizer='adam')

#  
model.fit(X_train, y_train, epochs=100, batch_size=4)

#  
pred = model.predict(X_test)

# bmi  
plt.figure(figsize=(10, 6))
plt.scatter(X_test['bmi'], y_test, label='test')  
plt.scatter(X_test['bmi'], pred, c='r', label='predict')  
plt.legend() 
plt.show()

# bp  
plt.figure(figsize=(10, 6))
plt.scatter(X_test['bp'], y_test, label='test')  
plt.scatter(X_test['bp'], pred, c='r', label='predict')  
plt.legend() 
plt.show()

# MSE, RMSE 
test_pred = model.predict(X_test)
train_pred = model.predict(X_train)

train_mse = mean_squared_error(y_train, train_pred) 
test_mse = mean_squared_error(y_test, test_pred)

print("train data set MSE :", train_mse)
print("test data set MSE :", test_mse)

print("train data set RMSE :", round(train_mse**0.5, 3))
print("test data set RMSE :", round(test_mse**0.5, 3))
```
